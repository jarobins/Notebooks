## #2 Find the Convolution of 2 Signals

Finding the convolution of two signals using built in formulas is rather easy and takes no real knowedge of the convolution process. Instead of doing the easy function, I have created a custom list instantiation from python than calculates the convolution on the fly. Using a predefined list(sig1) I can run a number of values through an equation and get the resulting convolution.  

List instantiation works in by populating a list array by defining a "for loop" that decides:
    1. How many items are in the list
    2. What it done to the items before adding them to the list

I use the equation:
    $$y[n] = 2x[n] - 3x[n-1] + 2x[n-2]$$

````python
sig1 = np.array([0,0,1,2,3,2,1,1,1,1,1])
sig3 = [2*sig1[n] - 3*sig1[n-1] + 2*sig1[n-2] for n in range(2, len(sig1))]
print "Signal 1:    ", sig1[2:]
print "Signal 2:     [2, -3, 2]"
print "Convolution: ", sig3
plt.stem(sig3)
plt.title("Convolution of Signal 1 and Signal 2")
plt.axis((-1, 9, -4, 4))
````

## #3 Find freq response

Using the difference equation: 
$$y[n]-\frac{4}{3}y[n-1]=x[n]-\frac{1}{2}x[n-1]$$

````python
s1 = signal.lti([1, -1.33], [1, -.5])
w, mag, phase = s1.bode()
x, impulse = s1.impulse()
y, step = s1.step()

plt.figure()
plt.semilogx(w, mag)      # Bode magnitude plot
plt.figure()
plt.semilogx(w, phase)    # Bode phase plot
plt.figure()
plt.semilogx(x, impulse)  # Impulse Response
plt.figure()
plt.semilogx(y, step)     # Step Response
plt.show()
````

## #4 Plot signal in time and in frequency

````python
x = np.arange(1001)
sig1 = np.sin(2*np.pi*(100/10000.0)*x) #1Hz Signal
sig2 = np.sin(2*np.pi*(200/10000.0)*x) #2Hz Signal
sig4 = sig1 + sig2

plt.plot(sig4)
plt.plot(sig1)
plt.plot(sig2)
plt.axis((0,100,-2,2))
````

````python
fft_output = np.fft.rfft(sig4)
magnitude_only = [np.sqrt(i.real**2 + i.imag**2)/len(fft_output) for i in fft_output]
frequencies = [(i*1.0/1001)*100 for i in range(1001/2+1)]
plt.plot(frequencies, magnitude_only)
plt.axis((0, 5, 0, 1.5))
````
